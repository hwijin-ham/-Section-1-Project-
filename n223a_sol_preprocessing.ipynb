### **2-1. Preprocessing for Linear Model**

```python
# cardinarlity ë†’ì€ ì»¬ëŸ¼ ì œê±°
import numpy as np

def engineer(df):
  cols = df
  label = cols.nunique()
  selected_cols = label[label <= 30].index.tolist()
  df = df[selected_cols]
  return df

train_lm = engineer(train_lm)
test_lm = engineer(test_lm)
```

```python
#behavioral ì»¬ëŸ¼ í•˜ë‚˜ë¡œ í•©ì¹˜ê¸° -> yes or noì˜ ì „ì²´ì ì¸ ë°©í–¥ì„±ì´ ê°™ìŒ

def behavioral(df):
  cols = df.columns
  cols_be = cols[cols.str.contains('behavioral_')]
  df['behavioral'] = 0
  for i in range(len(cols_be)):
    df['behavioral'] = df['behavioral'] + df[cols_be[i]]

  return df['behavioral']
```

```python
behavioral(train_lm)
```

```python
behavioral(test_lm)
```

```python
def dropbeh(df):
  cols = df.columns
  cols_be = cols[cols.str.contains('behavioral_')]
  df = df.drop(cols_be, axis=1)
  return df
```

```python
train_lm = dropbeh(train_lm)
```

```python
test_lm = dropbeh(test_lm)
```
