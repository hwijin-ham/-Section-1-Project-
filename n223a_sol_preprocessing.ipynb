## **Part 2. Linear Model**

### **2-1. Preprocessing for Linear Model**

- H1N1 데이터를 보고 선형 모델을 학습하기 전에 수행되어야 할 전처리와 그 이유를 서술해보세요.
 
- 지난 과제에 이어서 언급하신 여러가지 전처리를 수행해보고 비교해보세요.

- 가장 성능이 높은 모델로 테스트 세트에 대한 예측 결과를 캐글에 제출하세요.

```python
# cardinarlity 높은 컬럼 제거
import numpy as np

def engineer(df):
  cols = df
  label = cols.nunique()
  selected_cols = label[label <= 30].index.tolist()
  df = df[selected_cols]
  return df

train_lm = engineer(train_lm)
test_lm = engineer(test_lm)
```

```python
#behavioral 컬럼 하나로 합치기 -> yes or no의 전체적인 방향성이 같음

def behavioral(df):
  cols = df.columns
  cols_be = cols[cols.str.contains('behavioral_')]
  df['behavioral'] = 0
  for i in range(len(cols_be)):
    df['behavioral'] = df['behavioral'] + df[cols_be[i]]

  return df['behavioral']
```

```python
behavioral(train_lm)
```

```python
behavioral(test_lm)
```

```python
def dropbeh(df):
  cols = df.columns
  cols_be = cols[cols.str.contains('behavioral_')]
  df = df.drop(cols_be, axis=1)
  return df
```

```python
train_lm = dropbeh(train_lm)
```

```python
test_lm = dropbeh(test_lm)
```

```python
# rent_own_r - 응답자의 거주 현황 -> max 99 -> 아웃라이어 -> nan -> 중간값
out = train_lm[train_lm['rent_own_r'] > 2].index
```

```python
out = test_lm[test_lm['rent_own_r'] > 2].index
test_lm['rent_own_r'].iloc[out]=np.NaN
```

```python
features = train_lm.columns[train_lm.columns != 'vacc_h1n1_f']

X_train_lm = train_lm[features]
y_train_lm = train_lm[target]
X_test_lm = test_lm
```

```python
# ordinar encoding -> very low ~ very high와 같은 항목은 수치로 표현해야 유리하다고 생각.
from category_encoders import OrdinalEncoder
```

```python
# 명목형 변수 특성 -> onehotencoding
from category_encoders import OneHotEncoder

ohe = OneHotEncoder()
X_train_lm_en = ohe.fit_transform(X_train_lm)
X_test_lm_en = ohe.transform(X_test_lm)
```

```python
#결측치 -> 중간값으로 채운다 -> 이유가 있어서 응답하지 않은 설문으로 판단하여 영향을 미치지 않는 값 입력
from sklearn.impute import SimpleImputer

imputer = SimpleImputer()
X_train_lm_imp = imputer.fit_transform(X_train_lm_en)
X_test_lm_imp = imputer.transform(X_test_lm_en)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train_lm_imp, y_train_lm, 
                                                  random_state=2, train_size=0.8,
                                                  test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

linear = LinearRegression()
linear.fit(X_train, y_train)
y_pred = linear.predict(X_test_lm_imp)
```
